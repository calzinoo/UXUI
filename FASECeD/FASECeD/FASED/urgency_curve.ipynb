```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D

# Dati: (nome, persistenza, impatto)
issues = [
    ("Aggiungerei la funzione messaggi all'assistenza", 3.5, 2.5),
    ("Caratterizzare meglio la home", 4, 2),
    ("Correggere le maiuscole nei testi", 3.5, 1.5),
    ("Ridimensionare le icone", 4, 1),
]

# Estrai coordinate e etichette
x = np.array([i[1] for i in issues])
y = np.array([i[2] for i in issues])
labels = [i[0] for i in issues]
identifiers = [str(i+1) for i in range(len(issues))]

# Crea la figura
plt.figure(figsize=(8, 6))

# Disegna i punti con identificatori numerici
for xi, yi, id in zip(x, y, identifiers):
    plt.scatter(xi, yi, color='darkorange', s=80)
    plt.text(xi, yi, id, fontsize=9, ha='center', va='center', color='white', weight='bold')

# Curva di soglia orizzontale (es. impatto = 3)
threshold_x_vals = [0, 1, 2, 3, 4, 5]
threshold_y_vals = [4.5, 3.5, 2.5, 1.5, 0.5, 0.5]
plt.plot(threshold_x_vals, threshold_y_vals, color='black', linewidth=2, drawstyle='steps-post', label='Urgency Threshold')

# Etichette assi e titolo
plt.xlabel("Persistenza")
plt.ylabel("Impatto")
plt.title("Urgency Curve (Impatto vs Persistenza)")

# Imposta limiti
plt.xlim(0, 5)
plt.ylim(0, 5)

# Ticks ogni 0.5
plt.xticks(np.arange(0, 5.5, 0.5))
plt.yticks(np.arange(0, 5.5, 0.5))

# Griglia
plt.grid(True, linestyle='--', alpha=0.6)

# Legenda
legend_elements = [Line2D([0], [0], marker='o', color='w', label=f"{id}. {label}",
                          markerfacecolor='darkorange', markersize=10)
                   for id, label in zip(identifiers, labels)]
plt.legend(handles=legend_elements, title="Legenda", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
